# setup

```python
import os
import sys

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))

from alertbert.eval_grouping import roc_test_plot
from alertbert.model_eval_utils import load_ground_truth_label_vocabs

verbose = False
plot_macro_only = False
path = "../saved_models"
```

# original data config

```python
ait_ads_a_config = "original"
label_vocabs = load_ground_truth_label_vocabs(path=path, configuration=ait_ads_a_config)
model_ids = [
    "mlm_1l_4h_16d_zero_0k",
    f"mlm_1l_4h_16d_{ait_ads_a_config}_1_60k",
    f"mlm_1l_4h_16d_{ait_ads_a_config}_1_60k",
    f"mlm_1l_4h_16d_{ait_ads_a_config}_1_60k",
]
```

## timedelta

```python
roc_test_plot(
    model_id="timedelta",
    aitads_a_config="original",
    path=path,
    verbose=verbose,
    label_vocabs=label_vocabs,
    plot_macro_only=plot_macro_only,
)
```

## 4 attention heads

```python
roc_test_plot(
    model_id="mlm_1l_4h_16d_original_1_60k",
    aitads_a_config="original",
    layers=("embedding", "encoder"),
    dim_reduction=2,
    path=path,
    verbose=verbose,
    label_vocabs=label_vocabs,
    plot_macro_only=plot_macro_only,
)
```
